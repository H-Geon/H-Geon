```python
class MyError(Exception):
  pass
```

```python
def say_nick(nick):
  if nick =='바보':
    raise MyError()
  print(nick)
```

```python
try:
  say_nick("천사")
  say_nick("바보")
except MyError:
  print("허용되지 않는 별명입니다.")
```

```python
def positive(l):
  result=[]
  for i in l:
    if i>0:
      result.append(i)
  return result
```

```python
print(positive([1,-3,2,0,-5,6]))
```

```python
def positive(x):
  return x>0
```

```python
print(list(filter(positive,[1,-3,2,0,-5,6])))
```
